```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Сначала у нужных файлов мы уберем все лишние данные вручную и загрузим отредактированные файлы в папку __xvg files__

```python
def draw_graphic(path, name):
    data = pd.read_csv(path, header = None)
    data[name] = data[0].apply(lambda x: float(x.split()[1]))
    data.drop(0, axis = 1, inplace = True)
    plt.figure(figsize = [12, 8])
    plt.rcParams['font.size'] = 15
    ax = sns.lineplot(data = data)
    ax.set_xlabel('Time')
    ax.set_ylabel(name)
    plt.title(name)
    plt.savefig(name + '.png', format = 'png')
    plt.show()
```

```python
path = r'https://raw.githubusercontent.com/Sanches-wolf228/Bioinformatics-4/main/xvg%20files/potential.xvg'
name = 'Potential'
draw_graphic(path, name)
```

```python
path = r'https://raw.githubusercontent.com/Sanches-wolf228/Bioinformatics-4/main/xvg%20files/temperature.xvg'
name = 'Temperature'
draw_graphic(path, name)
```

```python
path = r'https://raw.githubusercontent.com/Sanches-wolf228/Bioinformatics-4/main/xvg%20files/rmsd.xvg'
name = 'RMSD'
draw_graphic(path, name)
```

```python
path = r'https://raw.githubusercontent.com/Sanches-wolf228/Bioinformatics-4/main/xvg%20files/gyrate.xvg'
data = pd.read_csv(path, header = None)
data['Radius'] = data[0].apply(lambda x: float(x.split()[2]))
data.drop(0, axis = 1, inplace = True)
data.index *= 10
plt.figure(figsize = [12, 8])
plt.rcParams['font.size'] = 15
ax = sns.lineplot(data = data)
ax.set_xlabel('Time')
ax.set_ylabel('Radius')
plt.title('Radius of gyration')
plt.savefig('gyrate.png', format = 'png')
plt.show()
```
